### Data Preprocessing

When imported into a Pandas dataframe using the default settings, the dataframe uses 2.6 GB of memory and contains 5819079 x 31 entries. Ultimately I reduced this to only 181.1 MB, a 93% reduction, using the following techniques:

1. Dropping columns that contained data irrelevant to predicting flight delays, such as flight number, aircraft registration number, flight distance, etc. 
2. Dropping flights to or from airports with fewer than 5 flights per day (ie, dropping outlier airports).
3. Downcasting each column's datatype to the most memory efficent datatype. By default, Pandas will import numerical data as either int64 or float64, and strings as objects. Large memory savings can be realized by downcasting one-hot encoded columns (such as CANCELLED or DIVERTED) from an 8-bit int64 to a 2-bit uint8. Columns containing departure and arrival delay times must be stored as float32, since they contain NaN values (for flights which were cancelled). Finally, massive savings can be realized by converting string columns (such as AIRLINE, DEPARTURE AIRPORT, and DESTINATION AIRPORT) as categories. These columns are perfect candidates for conversion to category data types because they contain few unique values compared to how many times each unique value appears.

Finally, I discovered that all of the flights that occured in October had corrupted arrival and departure airport data. Every other month used a three-letter IATA airport code (such as SFO or JFK), but the flights for October used a five digit Department of Transportation code (such as 14747). I created a dictionary that contained DOT Code : IATA Code mappings, which I then used to convert all of the airports in the dataframe that contained DOT Codes to their IATA equivalent.

### Web Scraping & Adding Weather Data

I used the Iowa State University's Mesonet service to download historical METAR (airport weather) data for the entire year for all of the flights in our dataset. The service helpfully provides a python script to scrape data from their site, which I modified for the purposes of this exploration. 

After downloading the data, I added weather data for each departure and arrival airport at the scheduled departure and arrival time to the dataframe.

The function I wrote to find and parse the correct weather data is below.

```python
def get_wx(airport, dt):
    
    # For a given airport and datetime, find the weather
    smallest_delta = datetime.timedelta(days=365)
    closest_time = 0
    airport_wx_df = airport_wx_dict[airport]
    for i in airport_wx_df['valid']:
        delta = abs(i-dt)
        if delta < smallest_delta: 
            smallest_delta = delta
            closest_time = i
        if delta > smallest_delta:
            break
    airport_wx = airport_wx_df[airport_wx_df['valid'] == closest_time]
    
    # Get the cloud ceiling (defined as the lowest Overcast or Broken cloud layer)
    if (airport_wx['skyc1'][0] == 'OVC') or (airport_wx['skyc1'][0] == 'BKN'):
        ceiling = pd.to_numeric(airport_wx['skyl1'], downcast='unsigned')
    elif airport_wx['skyc2'][0] == 'OVC' or airport_wx['skyc2'][0] == 'BKN':
        ceiling = pd.to_numeric(airport_wx['skyl2'], downcast='unsigned')
    elif airport_wx['skyc3'][0] == 'OVC' or airport_wx['skyc3'][0] == 'BKN':
        ceiling = pd.to_numeric(airport_wx['skyl3'], downcast='unsigned')
    else: ceiling = 25000
    
    # set missing wind values ('M') to 0 knots
    try:
        wind = pd.to_numeric(airport_wx['sknt'], downcast='unsigned')
    except:
        wind = 0  
    
    # set missing visibility values to 10 miles
    try:
        visibility = pd.to_numeric(airport_wx['vsby'], downcast='unsigned')
    except:
        visibility = 10
    
    # set missing precipitation values to 0 inches
    try:
        precip = pd.to_numeric(airport_wx['p01i'], downcast='unsigned')
    except: 
          precip=0
    
    return pd.DataFrame({'CEILING' : ceiling, 'VISIBILITY': visibility, 'WIND_SPEED': wind, 'PRECIPITATION': precip})    
```

### Modeling

The first models I created attempted to predict whether a particular flight would be cancelled or not. First, I one-hot encoded the categorical columns of AIRLINE, ORIGIN_AIRPORT, and ARRIVAL_AIRPORT. Then I split the dataframe into a training  subset and a testing subset because I want to avoid overfitting the model to just the training data and provide an independent dataset I can test against that the model has never seen before.

The best model was an XGBoost random forest model which had a precision of 0.84 and a recall of 0.11. Increasing the scale_pos_weight parameter led to a higher recall but a lower precision. Ultimately, I was reasonably satisfied with the default parameter of 1 to maximize the model's precision. In other words, I valued a low false-positive and a high false-negative rate over the inverse.
